```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Iris.csv")
data.head()
```

```python
data.info()
data.describe()
```

```python
data.hist(figsize=(10,8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Feature Distribution in Iris Dataset", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=data, palette='Set2')
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
sns.pairplot(data, hue='Species', diag_kind='hist', palette='husl')
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Species', y='PetalLengthCm', data=data, palette='pastel')
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
# Example of saving plots
plt.figure(figsize=(8,6))
sns.scatterplot(x='SepalLengthCm', y='PetalWidthCm', hue='Species', data=data)
plt.title("Saved Visualization Example")
plt.savefig("iris_scatterplot.png")
plt.close()

print("Plots generated and one saved as 'iris_scatterplot.png'")
```
